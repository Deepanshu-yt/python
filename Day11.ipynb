```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna(inplace=True)
print(df)

# fillna | fill null values
df.fillna(0,inplace=True)
print(df)
```
