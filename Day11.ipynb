```python
import matplotlib.pyplot as plt
Players = ["Neymar","Messi","Ronaldo","vini","Haland","Mbappe","Yamal"]
Goals = [344,520,330,234,322,233,123]

matches_played = [1, 2, 3, 4, 5, 6, 7]
winning_chances = [0.6, 0.7, 0.65, 0.8, 0.75, 0.9, 0.85]

assists = [5, 8, 7, 10, 9, 12, 11]
goals_per_match = [1.2, 1.5, 1.3, 1.8, 1.6, 2.0, 1.7]

player_names_for_bar = ["Messi", "Ronaldo", "Neymar", "Haland"]
player_ratings = [9.2, 9.0, 8.5, 7.6]
fig, axes = plt.subplots(2, 2, figsize=(15, 12))


axes[0, 0].pie(Goals, labels=Players, autopct='%1.1f%%', startangle=90, colors=["yellow","skyblue","red","orange","blue","violet","pink"])
axes[0, 0].set_title('Player Goals Distribution')
axes[0, 0].axis('equal')


axes[0, 1].plot(matches_played, winning_chances, marker='o', linestyle='-', color='skyblue')
axes[0, 1].set_title('Winning Chances Over Matches Played')
axes[0, 1].set_xlabel('Matches Played')
axes[0, 1].set_ylabel('Winning Chance')
axes[0, 1].grid(True, linestyle='--', alpha=0.7)

axes[1, 0].plot(matches_played, assists, marker='x', linestyle='--', color='green', label='Assists')
axes[1, 0].plot(matches_played, goals_per_match, marker='s', linestyle='-', color='purple', label='Goals Per Match')
axes[1, 0].set_title('Performance Metrics Over Matches')
axes[1, 0].set_xlabel('Matches Played')
axes[1, 0].set_ylabel('Count / Rate')
axes[1, 0].legend()
axes[1, 0].grid(True, linestyle='--', alpha=0.7)


axes[1, 1].bar(player_names_for_bar, player_ratings)
axes[1, 1].set_title('Top Player Ratings')
axes[1, 1].set_xlabel('Player')
axes[1, 1].set_ylabel('Rating')

plt.tight_layout()
plt.show()
```

### Cricket Player Statistics Visualization

Below, I will generate some dummy data for cricket players and then create a multi-panel plot to visualize various statistics, similar to the football player analysis.

```python
# --- Dummy Data for Cricket Statistics ---

# Data for Pie Chart (Player Runs)
cricket_players = ['Virat Kohli', 'Rohit Sharma', 'Steve Smith', 'Kane Williamson', 'Joe Root', 'Babar Azam', 'Ben Stokes']
runs_scored = [12898, 9825, 8479, 7921, 11000, 9500, 7500] # Example runs in ODI/Test combined

# Data for Line Chart 1 (Team Winning Percentage Over Matches)
matches_played_cricket = [1, 5, 10, 15, 20, 25, 30]
winning_percentage = [0.55, 0.60, 0.58, 0.65, 0.62, 0.70, 0.68] # Example team winning percentage

# Data for Multi-Line Chart (Matches vs Wickets & Batting Average)
wicket_taken = [10, 25, 40, 55, 70, 85, 100] # Example wickets taken by a bowler
batting_average = [35.5, 40.2, 38.1, 42.5, 39.8, 45.0, 43.2] # Example player batting average

# Data for Bar Chart (Top Player Ratings in Cricket)
top_cricket_players = ['Virat Kohli', 'Steve Smith', 'Babar Azam', 'Jasprit Bumrah']
cricket_ratings = [9.5, 9.3, 9.0, 8.8]
```

```python
import matplotlib.pyplot as plt

# Create a figure with a 2x2 grid of subplots
fig_cricket, axes_cricket = plt.subplots(2, 2, figsize=(15, 12)) # Adjust figsize for better layout

# Subplot 1 (Top-Left): Pie Chart for Player Runs
axes_cricket[0, 0].pie(runs_scored, labels=cricket_players, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'orange', 'silver'])
axes_cricket[0, 0].set_title('Cricket Player Runs Distribution')
axes_cricket[0, 0].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

# Subplot 2 (Top-Right): Line Chart for Team Winning Percentage
axes_cricket[0, 1].plot(matches_played_cricket, winning_percentage, marker='o', linestyle='-', color='teal')
axes_cricket[0, 1].set_title('Team Winning Percentage Over Matches')
axes_cricket[0, 1].set_xlabel('Matches Played')
axes_cricket[0, 1].set_ylabel('Winning Percentage')
axes_cricket[0, 1].grid(True, linestyle='--', alpha=0.7)

# Subplot 3 (Bottom-Left): Multi-Line Chart for Matches vs Wickets & Batting Average
axes_cricket[1, 0].plot(matches_played_cricket, wicket_taken, marker='x', linestyle='--', color='darkgreen', label='Wickets Taken')
axes_cricket[1, 0].plot(matches_played_cricket, batting_average, marker='s', linestyle='-', color='indigo', label='Batting Average')
axes_cricket[1, 0].set_title('Performance Metrics Over Matches (Cricket)')
axes_cricket[1, 0].set_xlabel('Matches Played')
axes_cricket[1, 0].set_ylabel('Count / Average')
axes_cricket[1, 0].legend()
axes_cricket[1, 0].grid(True, linestyle='--', alpha=0.7)

# Subplot 4 (Bottom-Right): Bar Chart for Top Cricket Player Ratings
axes_cricket[1, 1].bar(top_cricket_players, cricket_ratings, color='skyblue')
axes_cricket[1, 1].set_title('Top Cricket Player Ratings')
axes_cricket[1, 1].set_xlabel('Player')
axes_cricket[1, 1].set_ylabel('Rating')
axes_cricket[1, 1].set_ylim(0, 10) # Assuming ratings are out of 10

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()
plt.show()
```

### Business Data Visualization

Here's a visualization of some dummy business data, applying the same multi-panel plotting structure.

```python
# --- Dummy Data for Business Statistics ---

# Data for Pie Chart (Market Share by Product Category)
product_categories = ['Electronics', 'Clothing', 'Home Goods', 'Books', 'Groceries']
market_share_percentage = [35, 25, 20, 10, 10]

# Data for Line Chart 1 (Monthly Sales Trend)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_sales = [150000, 165000, 180000, 175000, 190000, 200000, 210000, 205000, 220000, 230000, 245000, 260000]

# Data for Multi-Line Chart (Quarterly Revenue vs. Expenses)
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
revenue = [500000, 550000, 600000, 680000]
expenses = [300000, 320000, 350000, 380000]

# Data for Bar Chart (Top 5 Product Profits)
top_products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
product_profits = [120000, 95000, 80000, 70000, 60000]
```

```python
import matplotlib.pyplot as plt

# Create a figure with a 2x2 grid of subplots
fig_business, axes_business = plt.subplots(2, 2, figsize=(15, 12)) # Adjust figsize for better layout

# Subplot 1 (Top-Left): Pie Chart for Market Share by Product Category
axes_business[0, 0].pie(market_share_percentage, labels=product_categories, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
axes_business[0, 0].set_title('Market Share by Product Category')
axes_business[0, 0].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

# Subplot 2 (Top-Right): Line Chart for Monthly Sales Trend
axes_business[0, 1].plot(months, monthly_sales, marker='o', linestyle='-', color='darkblue')
axes_business[0, 1].set_title('Monthly Sales Trend')
axes_business[0, 1].set_xlabel('Month')
axes_business[0, 1].set_ylabel('Sales ($)')
axes_business[0, 1].grid(True, linestyle='--', alpha=0.7)
axes_business[0, 1].ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis

# Subplot 3 (Bottom-Left): Multi-Line Chart for Quarterly Revenue vs. Expenses
axes_business[1, 0].plot(quarters, revenue, marker='x', linestyle='--', color='purple', label='Revenue')
axes_business[1, 0].plot(quarters, expenses, marker='s', linestyle='-', color='red', label='Expenses')
axes_business[1, 0].set_title('Quarterly Revenue vs. Expenses')
axes_business[1, 0].set_xlabel('Quarter')
axes_business[1, 0].set_ylabel('Amount ($)')
axes_business[1, 0].legend()
axes_business[1, 0].grid(True, linestyle='--', alpha=0.7)
axes_business[1, 0].ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis

# Subplot 4 (Bottom-Right): Bar Chart for Top 5 Product Profits
axes_business[1, 1].bar(top_products, product_profits, color='lightgreen')
axes_business[1, 1].set_title('Top 5 Product Profits')
axes_business[1, 1].set_xlabel('Product')
axes_business[1, 1].set_ylabel('Profit ($)')
axes_business[1, 1].ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()
plt.show()
```
