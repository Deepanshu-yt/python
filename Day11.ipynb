```python
# handle missing values
import numpy as np
import pandas as pd
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay']
}
df = pd.DataFrame(data)
print(df)
# null value filter | find out
print(df.isnull())

# sum of null
print(df.isnull().sum())

# dropna | drop complete row
df.dropna(inplace=True)
print(df)

# fillna | fill null values
df.fillna(0,inplace=True)
print(df)
```

```python
import matplotlib.pyplot as plt

Players = ["Neymar","Messi","Ronaldo","vini","Haland","Mbappe","Yamal"]
Goals = [344,520,330,234,322,233,123]


matches_played = [1, 2, 3, 4, 5, 6, 7]
winning_chances = [0.6, 0.7, 0.65, 0.8, 0.75, 0.9, 0.85]


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.pie(Goals, labels=Players, autopct='%1.1f%%', startangle=90, colors=["yellow","skyblue","red","orange","blue","violet","pink"])
plt.title('Player Goals Distribution')
plt.axis('equal')


plt.subplot(1, 2, 2)
plt.plot(matches_played, winning_chances, marker='*', linestyle='-', color='skyblue')
plt.title('Winning Chances Over Matches Played')
plt.xlabel('Matches Played')
plt.ylabel('Winning Chance')
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
Players = ["Neymar","Messi","Ronaldo","vini","Haland","Mbappe","Yamal"]
Goals = [344,520,330,234,322,233,123]

matches_played = [1, 2, 3, 4, 5, 6, 7]
winning_chances = [0.6, 0.7, 0.65, 0.8, 0.75, 0.9, 0.85]

assists = [5, 8, 7, 10, 9, 12, 11]
goals_per_match = [1.2, 1.5, 1.3, 1.8, 1.6, 2.0, 1.7]

player_names_for_bar = ["Messi", "Ronaldo", "Neymar", "Haland"]
player_ratings = [9.2, 9.0, 8.5, 7.6]
fig, axes = plt.subplots(2, 2, figsize=(15, 12))


axes[0, 0].pie(Goals, labels=Players, autopct='%1.1f%%', startangle=90, colors=["yellow","skyblue","red","orange","blue","violet","pink"])
axes[0, 0].set_title('Player Goals Distribution')
axes[0, 0].axis('equal')


axes[0, 1].plot(matches_played, winning_chances, marker='o', linestyle='-', color='skyblue')
axes[0, 1].set_title('Winning Chances Over Matches Played')
axes[0, 1].set_xlabel('Matches Played')
axes[0, 1].set_ylabel('Winning Chance')
axes[0, 1].grid(True, linestyle='--', alpha=0.7)

axes[1, 0].plot(matches_played, assists, marker='x', linestyle='--', color='green', label='Assists')
axes[1, 0].plot(matches_played, goals_per_match, marker='s', linestyle='-', color='purple', label='Goals Per Match')
axes[1, 0].set_title('Performance Metrics Over Matches')
axes[1, 0].set_xlabel('Matches Played')
axes[1, 0].set_ylabel('Count / Rate')
axes[1, 0].legend()
axes[1, 0].grid(True, linestyle='--', alpha=0.7)


axes[1, 1].bar(player_names_for_bar, player_ratings)
axes[1, 1].set_title('Top Player Ratings')
axes[1, 1].set_xlabel('Player')
axes[1, 1].set_ylabel('Rating')

plt.tight_layout()
plt.show()
```
