```python
# function calculator -> add, sub
def calculator( a , b , sign):
  if sign == '+':
    print('addition' , a+b)
  elif sign == '-':
    print('sub' , a-b)
  elif sign == '*':
    print('mul' , a*b)
  elif sign == '/':
    print('division' , a/b)

num1 = int(input("Enter a num"))
num2 = int(input("Enter a num"))
sign = input("Enter sign")
calculator(num1 , num2 , sign)
```

```python
#new method
def cal(n1,n2,opt):
  match opt:
    case"+":
      return n1+n2
    case"-":
      return n1-n2
    case"*":
      return n1*n2
    case"/":
      return n1/n2
    case _:
      return "Invalid input"

add=cal(3,4,'+')
sub=cal(3,4,'-')
mul=cal(3,4,'*')
div=cal(3,4,'/')

print(add)
print(sub)
print(mul)
print(div)
```

```python
# using eval function
def cal(*a):
  print(eval(f"{a[0]} {a[-1]} {a[1]}"))

cal(10,20,"-")
cal(10,20,"*")
cal(10,20,"+")
cal(10,20,"/")
cal(10,5,"**")
```
