```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private variable

    def deposit(self, amount):
        self.__balance += amount
        print(f"₹{amount} deposited")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"₹{amount} withdrawn")
        else:
            print("Insufficient Balance")

    def get_balance(self):
        return self.__balance


# Object Creation
acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(200)

print("Current Balance:", acc.get_balance())
```

```python
class Student:
    def __init__(self, name, marks):
        self.__name = name  # Public attribute
        self.__marks = marks  # Private attribute

    def get_name(self):
        return self.__name

    def get_marks(self):
        return self.__marks

    def set_marks(self, new_marks):
        if 0 <= new_marks <= 100:
            self.__marks = new_marks
            print(f"{self.__name}'s marks updated to {new_marks}")
        else:
            print("Invalid marks. Marks must be between 0 and 100.")

# Creating a student object
student1 = Student("Alice", 85)

# Accessing name and marks using getter methods
print(f"{student1.get_name()} has {student1.get_marks()} marks.")

# Trying to modify marks directly (will not work as expected due to name mangling)
# student1.__marks = 105
# print(student1.get_marks())

# Modifying marks using the setter method
student1.set_marks(92)
print(f"{student1.get_name()} now has {student1.get_marks()} marks.")

student1.set_marks(110) # Attempt to set invalid marks
```

```python
class Car:
    def __init__(self, make, model, year):
        self.__make = make  # Private attribute
        self.__model = model # Private attribute
        self.__year = year  # Private attribute

    def get_car_info(self):
        return f"Make: {self.__make}, Model: {self.__model}, Year: {self.__year}"

    def set_year(self, new_year):
        if 2000 <= new_year <= 2024:
            self.__year = new_year
            print(f"Year updated to {new_year}")
        else:
            print("Invalid year. Year must be between 2000 and 2024.")

# Creating an object
my_car = Car("Toyota", "Camry", 2020)

# Accessing info using a public method (getter)
print(my_car.get_car_info())

# Trying to directly access a private attribute (will cause an AttributeError if uncommented)
# print(my_car.__make)

# Modifying a private attribute using a public method (setter)
my_car.set_year(2022)
print(my_car.get_car_info())

my_car.set_year(1999) # This will show an error message
```
